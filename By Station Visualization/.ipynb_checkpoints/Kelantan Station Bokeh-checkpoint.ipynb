```python
import bokeh
import pandas_bokeh
from bokeh.plotting import figure, show
pandas_bokeh.output_notebook()
#pandas_bokeh.output_file("try.html")
```

```python
df1 = df1.sort_values(['Date'], ascending=[True])
```

```python
#create and configure the figure
dabongFig = figure(x_axis_type='datetime',
    plot_width = 450, plot_height = 350,
    title='Water Level in Sungai Galas, Dabong From 2016 to 2020',
    x_axis_label='Date', y_axis_label='Water Level (m)')



#Render the water level as step lines
dabongFig.line('Date', 'Sg. Galas Dabong (m)',
               line_width=2,
              color='blue', 
              source=df1)



show(dabongFig)
```

```python
df1.plot_bokeh.line(
    figsize=(800, 450),
    x="Date",
    y="Sg. Galas Dabong (m)",
    title="Water level in Sungai Galas, Dabong, Kelantan. From 2016 to 2020",
    xlabel="Date",
    ylabel="Water Level",
    ylim=(0, 55),
    panning=False,
    zooming=False)
```

```python
df1YearDabong = df1[(df1['Date'] > '2019-01-01') & (df1['Date'] < '2020-01-01')]
```

```python
df1YearDabong
```

```python
#Water level sungai galas in 2020
df1YearDabong.plot_bokeh.line(
    figsize=(800, 450),
    x="Date",
    y="Sg. Galas Dabong (m)",
    title="Water level at Sungai Galas, Dabong, Kelantan in 2020",
    xlabel="Date",
    ylabel="Water Level",
    ylim=(0, 55),
    panning=False,
    zooming=False)
```
